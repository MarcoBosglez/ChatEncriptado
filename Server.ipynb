```python
import time, socket, sys, random
print('Setup Server...')
time.sleep(1)
#Get the hostname, IP Address from socket and set Port
soc = socket.socket()
host_name = socket.gethostname()
ip = socket.gethostbyname(host_name)
port = 1234
soc.bind((host_name, port))
print(host_name, '({})'.format(ip))
name = input('Enter name: ')
soc.listen(1) #Try to locate using socket
print('Waiting for incoming connections...')
connection, addr = soc.accept()
print("Received connection from ", addr[0], "(", addr[1], ")\n")
print('Connection Established. Connected From: {}, ({})'.format(addr[0], addr[0]))
#get a connection from client side
client_name = connection.recv(1024)
client_name = client_name.decode()
print(client_name + ' has connected.')
connection.send(name.encode())

def Crypt(string,key,encrypt=1):
    """
    Key
    """
    random.seed(key)
   
    alphabet = 2 * "AaBbCcDdEeFfGgHhIiJjKkLlMmNnOoPpQqRrSsTtUuVvWwXxYyZz Â¿?,123456789"
    
    lenalpha = int(len(alphabet)/2)
    if encrypt:
        
        return ''.join([alphabet[alphabet.index(string[p]) + lenalpha - int(lenalpha * random.random())] for p in range(len(string))])
    else:
        
        return ''.join([alphabet[alphabet.index(string[p]) - lenalpha + int(lenalpha * random.random())] for p in range(len(string))])
key = input("My key ")
while True:
    message = input('Me > ')
    if message == 'BYE':
        message = 'Good Night...'
        connection.send(message.encode())
        print("\n")
        break
    var=1
    while var==1:
        encrypt_question = input("Press 1 to encrypt ")
        if encrypt_question in ['1','0']:
            finished_with_the_user = True
            encrypt_question= int(encrypt_question)
            print(' Tu mensaje fue: ', message, 'Tu nuevo mensaje es: ', Crypt(message, key, encrypt_question))
            connection.send(Crypt(message, key, encrypt_question).encode())
            break
        else:
            print("Please input a valid number. ")
    message = connection.recv(1024)
    message = message.decode()
    print(client_name, '>', message)
    var=1
    while var==1:
        encrypt_question = input('Press 0 to decrypt ')
        if encrypt_question in ['1','0']:
            finished_with_the_user = True
            encrypt_question= int(encrypt_question)
            print(Crypt(message, key, encrypt_question))
            break
        else:

            print("Please input a valid number. ")
```
