```python
import time, socket, sys, random

print('Client Server...')
time.sleep(1)
#Get the hostname, IP Address from socket and set Port
soc = socket.socket()
shost = socket.gethostname()
ip = socket.gethostbyname(shost)
#get information to connect with the server
print(shost, '({})'.format(ip))
server_host = input('Enter server\'s IP address:')
name = input('Enter Client\'s name: ')
port = 1234
print('Trying to connect to the server: {}, ({})'.format(server_host, port))
time.sleep(1)
soc.connect((server_host, port))
print("Connected...\n")
soc.send(name.encode())
server_name = soc.recv(1024)
server_name = server_name.decode()
print('{} has joined...'.format(server_name))
def Crypt(string,key,encrypt=1):
    """
    Key
    """
    random.seed(key)
   
    alphabet = 2 * "AaBbCcDdEeFfGgHhIiJjKkLlMmNnOoPpQqRrSsTtUuVvWwXxYyZz Â¿?,123456789"
    
    lenalpha = int(len(alphabet)/2)
    if encrypt:
        
        return ''.join([alphabet[alphabet.index(string[p]) + lenalpha - int(lenalpha * random.random())] for p in range(len(string))])
    else:
        
        return ''.join([alphabet[alphabet.index(string[p]) - lenalpha + int(lenalpha * random.random())] for p in range(len(string))])
key = input("My key ")

while True:
    message = soc.recv(1024)
    message = message.decode()
    print(server_name, '>', message)
    var=1
    while var==1:
        encrypt_question = input('Press 0 to decrypt ')
        if encrypt_question in ['1','0']:
            finished_with_the_user = True
            encrypt_question= int(encrypt_question)
            print(Crypt(message, key, encrypt_question))
            break
        else:

            print("Please input a valid number. ")
    message = input('Me > ')
    if message == "BYE":
        message = "Leaving the Chat room"
        soc.send(message.encode())
        print("\n")
        break

    var=1
    while var==1:
        encrypt_question = input('Press 1 to encrypt ')
        if encrypt_question in ['1','0']:
            finished_with_the_user = True
            encrypt_question= int(encrypt_question)
            print(' Tu mensaje fue: ', message, 'Tu nuevo mensaje es: ', Crypt(message, key, encrypt_question))
            soc.send(Crypt(message, key, encrypt_question).encode())
            break
        else:
            print("Please input a valid number. ")
    

```
